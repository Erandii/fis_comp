$$
y''+w_0^2sen(y)=g(t)\\
y'=u\\
u'=-w_0^2sen(y)+g(t)\\
$$

```python
from pylab import*
%run "integradores_pro.ipynb"
%matplotlib inline
```

```python
def osc(y,t,w02=1.):
    return array([y[1],-w02*sin(y[0])])
```

```python
tiempos,sol=runge_kutta4(osc,[1.,0.],[0.,10.])
plot(sol[:,0],sol[:,1])
zoom=2
xlim([-zoom, zoom]),ylim([-zoom, zoom])
axes().set_aspect('equal'),grid(True)
show()
```

Estamos graficando el espacio fase: $\hat{\theta} = u$ vs $\theta=y$.

```python
for i in range(10):
    y0,u0=pi*(2*rand()-1),0 #4*(2*rand()-1)
    t,sol=runge_kutta4(osc,[y0,u0],[0.,10.])
    plot(sol[:,0],sol[:,1])
zoom=pi
xlim([-zoom, zoom]),ylim([-zoom, zoom])
axes().set_aspect('equal'),grid(True)
show()
```

NOTA: las lneas que no se unen en un rombo deberan de salir en x=4 y entrar en x=-4

```python
def osc1(y,t,w02=1.):
    return array([y[1],-w02*sin(y[0])+0.1*sin(t)])
```

```python
for i in range(3):
    y0,u0=pi*(2*rand()-1),0 #4*(2*rand()-1)
    t,sol=runge_kutta4(osc1,[y0,u0],[0.,100.])
    plot(sol[:,0],sol[:,1])
zoom=pi
xlim([-zoom, zoom]),ylim([-zoom, zoom])
axes().set_aspect('equal'),grid(True)
show()
```

```python
y0,u0=1.2,0
t,sol=runge_kutta4(osc1,[y0,u0],[0.,10000.])
plot(sol[:,0],sol[:,1])
zoom=pi
xlim([-zoom, zoom]),ylim([-zoom, zoom])
axes().set_aspect('equal'),grid(True)
show()
```

NOTA: La trayectoria queda restringida a cierta rea del espacio fase

```python
y0,u0=1.2,1
t,sol=runge_kutta4(osc1,[y0,u0],[0.,2000.])
plot(sol[:,0],sol[:,1])
y0,u0=1.2,0
t,sol=runge_kutta4(osc1,[y0,u0],[0.,2000.])
plot(sol[:,0],sol[:,1])
zoom=pi
xlim([-zoom, zoom]),ylim([-zoom, zoom])
axes().set_aspect('equal'),grid(True)
show()
```
