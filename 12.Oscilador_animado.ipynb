# Oscilador animado

```python
from pylab import*
from scipy.integrate import*
%matplotlib inline
```

```python
#y_v[0]=y, y_v[1]=v
def osc(y_v,t, omega2=1.,mu=0):
    dy = y_v[1]
    dv = -omega2*y_v[0] - mu*y_v[1]
    return array([dy,dv])
```

```python
t = linspace(0,50,500)
inicio = [0.5,0.8]
sol = odeint(osc,inicio,t,args=(1.,0.1))
xlim([-1,1])
ylim([-1,1])
plot(sol[:,0],sol[:,1])
show()
```

```python
from JSAnimation import IPython_display
from matplotlib import animation
```

```python
fig = figure()
# el codigo 0xb3.... es la identificación de la parte de la memoria donde esta guadada la imágen
```

```python
figura = figure()
ejes = axes(xlim=(-1,1),ylim=(-1,1) )

trayectoria, = ejes.plot([],[], lw=2) 
punto, = ejes.plot([],[], lw=2)
#esto es algo que asigna datos a una gráfica.
#con ello, hacemos que en el mismo entorno se escriban distintas cosas. 

def init():
    #queremos que los datos sean nulos, por eso estamos escribiendo ([],[])
    #este programa sirve para que cada vez  
    trayectoria.set_data([],[])
    punto.set_data([],[])
    return trayectoria,punto,

def el_movimiento(i):
    y,v=sol[:,0],sol[:,1]
    trayectoria.set_data(y[:i+1],v[:i+1])
    punto.set_data(y[i],v[i])
    return trayectoria,punto,

animation.FuncAnimation(figura,el_movimiento,init_func=init,frames=100)
#ésta función necesita tres argumentos: entorno, con quién hago dibujos, con que hago borron y cuenta nueva.
#lo que nos está dando esta función como "resultado" es un java
```
